```python
import pandas as pd
grid ="3"
```

```python
Results = pd.read_csv("grid_"+grid+".csv")
Results = Results.sort_values(by=["mean_test_score"])
```

```python
_x = list(Results["param_c1"])
_y = list(Results["param_c2"])
_f1 = list(Results["mean_test_score"])
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.colors as colors
plt.style.use('ggplot')

fig = plt.figure()
fig.set_size_inches(12, 12)
ax = plt.gca()
ax.set_xlabel('L1')
ax.set_ylabel('L2')

ax.set_title("Hyperparameter Grid Search Results")

norm_1 = colors.LogNorm(vmin=min(_f1),vmax=max(_f1))

ax.scatter(_x, _y, c=_f1, s=100, alpha=1, edgecolors=[0,0,0], norm=norm_1, cmap='viridis')
c1 = Results.sort_values(by=["rank_test_score"])["param_c1"].iloc[0]
c2 = Results.sort_values(by=["rank_test_score"])["param_c2"].iloc[0]
ax.scatter(c1,c2,color="red", s=100)
#fig.colorbar(mpl.cm.ScalarMappable(norm=norm_1),ax=ax)

sm = mpl.cm.ScalarMappable(cmap='viridis', norm=norm_1)
sm._A = []
sm.get_array
fig.colorbar(sm, ax=ax, ticks=[min(_f1), max(_f1)])
```

```python
fig.savefig('grid_'+grid+'.png')
```
